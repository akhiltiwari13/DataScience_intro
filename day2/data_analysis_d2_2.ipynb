```python
g=sns.countplot(df['Gender'])

plt.xlabel('Gender of Passenger')
plt.ylabel('Count of Passenger')
plt.yticks([0,500,1000])
plt.title('Gender Count')

plt.rc('font',size=12)
plt.rc('axes',titlesize=40)
plt.rc('ytick',labelsize=30)

plt.show(g)
type(g)
```

```python
g=sns.distplot(df['Age'])

plt.xlabel('Age of Passenger')
plt.ylabel('Densitiy')
plt.xticks([x for x in range(0,80,10)])
plt.title('Distplot')

plt.show(g)
type(g)
```
