```python
#1. Load 300 rows from titanic data set

import pandas as pd
df = pd.read_csv('../data_files/titanic.csv', nrows=300)
df
```

```python
# 2. Group the data on Pclass,Embarked and Gender column

grp = df.groupby(['Pclass','Embarked', 'Gender'])
grp.groups
# grp.get_group((3,'S','male'))
```

```python
# 3. Drop data of all female passengers in Pclass 3 who embarked from 'S'

condition = (df['Embarked'] == 'S') & (df['Gender']=='female') & (df['Pclass'] == 3)
# df.drop(df[condition],axis=0,inplace=True) #doesn't work?
# Since I couldn't get drop to work in this case, assigning data_frame to a filtered version of itself.
df=df[~condition]
df
```

```python
# 4. Show the statistical data for all passengers who travelled from Pclass 2
df[df['Pclass']==2].describe()
```

```python
# 5. Calculate the total fare collected from male passengers in Pclass 3
df[df['Pclass']==3][['Fare']].sum()
```

```python
# 6. Show the Names of passengers who paid a fare of above 100 if they embarked from C

condition =(df['Fare'] > 100)& (df['Embarked']== 'C')
df[condition]['Name']
```

```python
# 7. Create a column next to Name column which shows the fare paid by them if 18% service tax was applied on it

def chargeServiceTax(x):
    return 1.18*x

df['TAXED_FARE']= df['Fare'].apply(chargeServiceTax)
df
```
